```python
import pandas as pd
import csv
import plotly.express as px

stars_w_gravity_df = pd.read_csv('https://drive.google.com/uc?id=1hf_L-8BAQTp4bSZdlMs8jQPnGpmN9Siw') #https://drive.google.com/file/d/1hf_L-8BAQTp4bSZdlMs8jQPnGpmN9Siw/view?usp=drive_link

star_radius_list = sorted(stars_w_gravity_df["Radius"].tolist())
star_mass_list = sorted(stars_w_gravity_df["Mass"].tolist())

print("Radius list : " , star_radius_list)
print("Mass list : " , star_mass_list )
```

```python
scatter = px.scatter(x = star_radius_list , y = star_mass_list)
scatter.show()
```
